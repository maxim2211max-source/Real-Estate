```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
path = "/home/adminadmin/rent_real_estate/_data.csv"
df = pd.read_csv(path)
print("Загружено:", df.shape)  # Проверь размер


rename_dict = {
    'Окна': 'windows',
    'Санузел': 'bathroom',
    'Можно с детьми/животными': 'allowed_with_children_pets',
    'Дополнительно': 'additional',
    'Название ЖК': 'residential_complex_name',
    'Серия дома': 'house_series',
    'Высота потолков, м': 'ceiling_height_m',
    'Лифт': 'elevator',
    'Мусоропровод': 'garbage_chute',
    'Ссылка на объявление': 'listing_link'
}
df.rename(columns=rename_dict, inplace=True)



df.drop(["house_series", "residential_complex_name"], axis=1, inplace=True)


if 'ceiling_height_m' in df.columns:
    median_height = df['ceiling_height_m'].median()
    df['ceiling_height_m'].fillna(median_height, inplace=True)
    

df.to_csv('data.csv', index=False)  

```
