```python
# ПРЕДВАРИТЕЛЬНАЯ ПОДГОТОВКА ДАННЫХ ДЛЯ ПРОВЕРКИ ВЫБРОСОВ
print("ПОДГОТОВКА ДАННЫХ ДЛЯ ПРОВЕРКИ ВЫБРОСОВ:")
print("=" * 25)

# Создаем необходимые столбцы для анализа выбросов
df['price'] = df['Цена'].str.extract('(\d+)').astype(float)
df['total_area'] = df['Площадь, м2'].astype(str).str.split('/').str[0].str.extract('(\d+\.?\d*)').astype(float)
df['rooms_clean'] = df['Количество комнат'].astype(str).str.extract('(\d+)').astype(float)

# Извлекаем количество лифтов
if 'Лифт' in df.columns:
    df['lift_count'] = df['Лифт'].str.extract('(\d+)').astype(float)

print("Созданные столбцы для анализа:")
print(f"  price: {df['price'].notna().sum()} значений")
print(f"  total_area: {df['total_area'].notna().sum()} значений") 
print(f"  rooms_clean: {df['rooms_clean'].notna().sum()} значений")
if 'lift_count' in df.columns:
    print(f"  lift_count: {df['lift_count'].notna().sum()} значений")

# ПРОВЕРКА И ФИЛЬТРАЦИЯ ВЫБРОСОВ
print("\nПРОВЕРКА НА ВЫБРОСЫ:")
print("=" * 25)

# 1. Анализ выбросов в цене
print("1. Анализ выбросов в цене:")
Q1_price = df['price'].quantile(0.25)
Q3_price = df['price'].quantile(0.75)
IQR_price = Q3_price - Q1_price
lower_bound_price = max(0, Q1_price - 1.5 * IQR_price)  # Цена не может быть отрицательной
upper_bound_price = Q3_price + 1.5 * IQR_price

outliers_price = df[(df['price'] < lower_bound_price) | (df['price'] > upper_bound_price)]
print(f"   Выбросы по цене: {len(outliers_price)} записей")
print(f"   Диапазон нормальных цен: {lower_bound_price:,.0f} - {upper_bound_price:,.0f} руб.")
if len(outliers_price) > 0:
    print("   Примеры выбросов по цене:")
    print(outliers_price[['price', 'total_area', 'rooms_clean']].head())

# 2. Анализ выбросов в площади
print("\n2. Анализ выбросов в площади:")
Q1_area = df['total_area'].quantile(0.25)
Q3_area = df['total_area'].quantile(0.75)
IQR_area = Q3_area - Q1_area
upper_bound_area = Q3_area + 1.5 * IQR_area

outliers_area = df[df['total_area'] > upper_bound_area]
print(f"   Выбросы по площади: {len(outliers_area)} записей")
print(f"   Максимальная нормальная площадь: {upper_bound_area:.1f} м²")
if len(outliers_area) > 0:
    print("   Примеры выбросов по площади:")
    print(outliers_area[['total_area', 'price', 'rooms_clean']].head())

# 3. Анализ выбросов в количестве комнат
print("\n3. Анализ выбросов в количестве комнат:")
room_outliers = df[(df['rooms_clean'] == 0) | (df['rooms_clean'] > 10)]
print(f"   Подозрительные значения комнат: {len(room_outliers)} записей")
if len(room_outliers) > 0:
    print("   Примеры подозрительных значений комнат:")
    print(room_outliers[['rooms_clean', 'price', 'total_area']].head())

# 4. Анализ выбросов в лифтах
print("\n4. Анализ выбросов в лифтах:")
if 'lift_count' in df.columns:
    lift_outliers = df[df['lift_count'] > 10]  # Более 10 лифтов - подозрительно
    print(f"   Подозрительные значения лифтов: {len(lift_outliers)} записей")
    if len(lift_outliers) > 0:
        print("   Примеры выбросов по лифтам:")
        print(lift_outliers[['Лифт', 'lift_count', 'price', 'total_area']].head())

# 5. Анализ выбросов в высоте потолков
print("\n5. Анализ выбросов в высоте потолков:")
if 'Высота потолков, м' in df.columns:
    ceiling_outliers = df[(df['Высота потолков, м'] < 2) | (df['Высота потолков, м'] > 5)]
    print(f"   Подозрительные значения высоты потолков: {len(ceiling_outliers)} записей")
    if len(ceiling_outliers) > 0:
        print("   Примеры выбросов по высоте потолков:")
        print(ceiling_outliers[['Высота потолков, м', 'price', 'total_area']].head())

# ВИЗУАЛИЗАЦИЯ ВЫБРОСОВ
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.boxplot(df['price'].dropna())
plt.title('Выбросы в цене')

plt.subplot(2, 3, 2)
plt.boxplot(df['total_area'].dropna())
plt.title('Выбросы в площади')

plt.subplot(2, 3, 3)
plt.boxplot(df['rooms_clean'].dropna())
plt.title('Выбросы в количестве комнат')

if 'lift_count' in df.columns and df['lift_count'].notna().sum() > 0:
    plt.subplot(2, 3, 4)
    plt.boxplot(df['lift_count'].dropna())
    plt.title('Выбросы в количестве лифтов')

if 'Высота потолков, м' in df.columns and df['Высота потолков, м'].notna().sum() > 0:
    plt.subplot(2, 3, 5)
    plt.boxplot(df['Высота потолков, м'].dropna())
    plt.title('Выбросы в высоте потолков')

plt.tight_layout()
plt.show()

# ФИЛЬТРАЦИЯ ВЫБРОСОВ
print("\nФИЛЬТРАЦИЯ ВЫБРОСОВ:")
print("=" * 25)

# Сохраняем исходное количество записей
original_count = len(df)

# Создаем маску для фильтрации
mask = (
    (df['price'] >= lower_bound_price) & 
    (df['price'] <= upper_bound_price) &
    (df['total_area'] <= upper_bound_area) &
    (df['rooms_clean'] >= 1) & 
    (df['rooms_clean'] <= 10)
)

# Дополнительная фильтрация лифтов
if 'lift_count' in df.columns:
    mask = mask & (df['lift_count'] <= 10)

# Применяем фильтрацию
df_clean = df[mask]

print(f"Исходное количество записей: {original_count}")
print(f"После фильтрации выбросов: {len(df_clean)}")
print(f"Удалено записей: {original_count - len(df_clean)} ({((original_count - len(df_clean))/original_count)*100:.1f}%)")

# РАБОТАЕМ С ОЧИЩЕННЫМИ ДАННЫМИ
df = df_clean.copy()

print("\n" + "="*50)
print("ДАЛЬНЕЙШИЕ АНАЛИЗЫ БУДУТ НА ОЧИЩЕННЫХ ДАННЫХ")
print("="*50)
```

```python
def extract_area(area_str):
    if pd.isna(area_str):
        return np.nan
    if isinstance(area_str, str):
        parts = area_str.split('/')
        if parts:
            try:
                return float(parts[0])
            except:
                return np.nan
    return np.nan

df['area_numeric'] = df['Площадь, м2'].apply(extract_area)

plt.figure(figsize=(12, 6))
plt.scatter(df['area_numeric'], df['price'], alpha=0.6, color='purple')
plt.xlabel('Площадь (м²)')
plt.ylabel('Цена (руб.)')
plt.title('Зависимость цены от площади')
plt.grid(True, alpha=0.3)

plt.savefig('img/price_distribution')
plt.show()
```
