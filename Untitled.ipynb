```python
def extract_price(text):
    match = re.search(r'(\d+\.?\d*)', str(text))  # Ищет первое число
    return float(match.group(1)) if match else None

df['Цена'] = df['Цена'].apply(extract_price)  # Новая колонка с числами
```

```python
print("Анализ зависимости цены от наличия мусропровода.")

garbage = df.groupby("Мусоропровод")["Цена"].median()

plt.figure(figsize = (9, 6))
garbage.plot(kind="bar", color = "blue")

plt.xlabel("Наличие мусоропровода")
plt.ylabel("Price")
plt.title("Анализ зависимости цены от наличия мусропровода.")
plt.show()
```

```python
print("Анализ зависимости цены от наличия/кол-ва лифтов.")

elevator = df.groupby("Лифт")["Цена"].median()
elevator = elevator.sort_values(ascending=False)

plt.figure(figsize = (9, 6))
elevator.plot(kind="bar", color = "blue")

plt.xlabel("Вид лифта")
plt.ylabel("Price")
plt.title("Анализ зависимости цены от наличия/кол-ва лифтов.")
plt.show()
```

```python
print("Зависимость цены от высоты потолков")
ceiling = df.groupby("Высота потолков, м")["Цена"].median()
ceiling = ceiling.sort_values(ascending = False)

plt.figure(figsize = (16, 8))
ceiling.plot(kind = "bar", color = "red")
plt.xlabel("level_ceiling")
plt.ylabel("Price")
plt.tight_layout()  
plt.show()
```

```python
key_features = ["Мебель в комнатах", "Мебель на кухне", "Ванна", "Душ", "Кондиционер", "Холодильник", "Стиральная машина"]  # Список признаков

# Проверка первых строк (показываем "Дополнительно" для примера)
print(df[["Дополнительно"]].head())

# Анализ медианных цен по наличию признаков (без создания колонок)
print("\nМедианные цены по наличию признаков:")
for feature in key_features:
    median_with = df[df["Дополнительно"].str.contains(feature, na=False)]["Цена"].median()
    median_without = df[~df["Дополнительно"].str.contains(feature, na=False)]["Цена"].median()
    count_with = df["Дополнительно"].str.contains(feature, na=False).sum()
    print(f"{feature}: С наличием — {median_with:.0f} руб (кол-во: {count_with}), Без — {median_without:.0f} руб")

# Подготовка данных для визуализации
medians = []
for feature in key_features:
    medians.append(df[df["Дополнительно"].str.contains(feature, na=False)]["Цена"].median())
    medians.append(df[~df["Дополнительно"].str.contains(feature, na=False)]["Цена"].median())

labels = [f"{f} (есть)" for f in key_features] + [f"{f} (нет)" for f in key_features]

# Визуализация бар-чарта
plt.figure(figsize=(12, 6))
plt.bar(labels, medians, color=['green' if 'есть' in l else 'red' for l in labels])
plt.title("Медианные цены по наличию удобств")
plt.ylabel("Медианная цена (руб/мес)")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()


```
